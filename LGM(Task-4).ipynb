#  Exploratory Data Analysis on Dataset - Terrorism 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='ISO-8859-1')
```

```python
df.info()
```

```python
#the columns heading not make any sense,so we will rename it for better understanding.
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType',
                     'target1':'Target','success':'Success','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
list(df.columns)
```

Not all 135 columns are necessary here because 99% of them contains only null values.
So we only use only necessary columns to perform Exploratory Data Analysis(EDA)

```python
new_cols = ['Year','Month','Day','Country','State','Region','latitude','longitude','AttackType','Success','Target','Killed','Wounded','Group','Target_type',
                   'Weapon_type']

df = df[new_cols]
```

```python
df.head()
```

```python
df.shape
```

```python
for i in df.columns:
    print(df[i].isnull().sum())
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.shape
```

```python
df.describe()
```

```python
df.corr()
```

```python
#Correlation Analysis
plt.figure(figsize=(15,10))

sns.heatmap(np.round(df.corr(),4), annot=True, cmap = 'viridis')
plt.xticks(rotation = 90)
plt.yticks(rotation = 0)
```

## Data Visualization

```python
plt.figure(figsize=(25,12))
sns.countplot(df['Year'], color='b', edgecolor='k')
plt.xlabel('Attack Year',fontweight='bold',fontsize=20)
plt.ylabel('Number Of Attack',fontweight='bold',fontsize = 20)
plt.title('Number OF Terror Attack Each Year', fontweight='bold',fontsize=20)
plt.xticks(rotation = 90)
plt.show()
```

### Insights
There are a large number of Terrorists Activities in the year 2014 an 2015

```python
plt.figure(figsize=(20,14))
sns.countplot(x=df['Year'], hue='Success', data=df
              , edgecolor = 'black')
plt.xlabel('Attack Year',fontweight='bold')
plt.ylabel('Number Of Attack',fontweight='bold')
plt.title('Number OF Terror Attack Each Year', fontweight='bold',fontsize=20)
plt.xticks(rotation = 90)
plt.legend(title='Attacks', loc='upper left', labels=['Success Attack', 'Unsuccess Attack'], fontsize=12)
```

### Insights
Thge most successful Attacks done by terrorist is in the year 2016.

```python
plt.figure(figsize=(14,10))
sns.countplot(df['AttackType'], order=df['AttackType'].value_counts().index, edgecolor='k',palette='Set2')
plt.xlabel('Attack Type',fontweight='bold')
plt.ylabel('Number Of Attack',fontweight='bold')
plt.title('Type OF Attack', fontweight='bold',fontsize=20)
plt.xticks(rotation = 60)
plt.show()
```

### Insights
The most used type of attack by the terrorists is the 'Bombing/Explosion'

```python
plt.figure(figsize=(14,8))
sns.barplot(y=df['Group'].value_counts()[1:10].index,x=df['Group'].value_counts()[1:10].values,
           palette='Set1',edgecolor='k')
plt.xlabel('Number of Attack',fontweight='bold',fontsize=14)
plt.ylabel('Name of Terrorist Group', fontweight='bold',fontsize=14)
plt.title('Most Active Terrorist Organizations', fontweight='bold',fontsize=18)
plt.show()
```

### Taliban and ISIL are the Most Active Terrorist Organisation

```python
# What may be the main target of the Terrorist?
plt.figure(figsize=(14,8))
sns.countplot(df['Target_type'], order=df['Target_type'].value_counts().index, edgecolor='k',palette='copper')
plt.xlabel('Target type',fontweight='bold')
plt.ylabel('Number Of Attack',fontweight='bold')
plt.title('Target type', fontweight='bold',fontsize=20)
plt.xticks(rotation = 90)
```

### Insights
The main target of the Terrorists is Private Citizens and Military.
